```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv("Unemployment in India - Unemployment in India.csv")
df2 = pd.read_csv("Unemployment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv")

# Display first few rows
print(df1.head())
print(df2.head())

# Plot unemployment rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df2)
plt.xticks(rotation=45)
plt.title("Unemployment Rate Over Time in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Drop rows with any missing values
df1_clean = df1.dropna()
df2_clean = df2.dropna()

# Check for missing values after dropping
print(df1_clean.isnull().sum())
print(df2_clean.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Ensure 'Date' columns are in datetime format
df1_clean['Date'] = pd.to_datetime(df1_clean['Date'], errors='coerce')
df2_clean['Date'] = pd.to_datetime(df2_clean['Date'], errors='coerce')

# Plot the Unemployment Rate Over Time (for df1_clean and df2_clean)
plt.figure(figsize=(12,6))

# Check if the Date columns are not empty and are in proper datetime format
if not df1_clean['Date'].isnull().all() and not df2_clean['Date'].isnull().all():
    plt.plot(df1_clean['Date'], df1_clean['Estimated Unemployment Rate (%)'], label='Dataset 1', color='blue')
    plt.plot(df2_clean['Date'], df2_clean['Estimated Unemployment Rate (%)'], label='Dataset 2', color='orange')
else:
    print("Error: One of the 'Date' columns is invalid.")

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
